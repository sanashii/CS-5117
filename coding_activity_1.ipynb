```python
import numpy as np
import matplotlib.pyplot as plt

def x_t(t):
    return 9 * np.cos(200 * np.pi * t + 0.4 * np.pi)

# -0.02 <= t <= 0.02
t_min = -0.02
t_max = 0.02
t_continuous = np.linspace(t_min, t_max, 1000) # for line tracing

sampling_intervals = [0.0075, 0.005, 0.0005]
sampling_labels = ['T_s = 0.0075 s', 'T_s = 0.005 s', 'T_s = 0.0005 s']


plt.figure(figsize=(10, 8))

for i, Ts in enumerate(sampling_intervals):
    t_values = np.arange(t_min, t_max, Ts)
    x_values = x_t(t_values)
    x_continuous = x_t(t_continuous)
    
    plt.subplot(3, 1, i + 1)
    plt.plot(t_continuous, x_continuous, label="Continuous Trace")
    plt.stem(t_values, x_values, basefmt=" ", label="Sampled Points")  
    plt.title(f'Sampled Signal with {sampling_labels[i]}')
    plt.xlabel('Time (s)')
    plt.ylabel('x(t)')
    plt.legend()

plt.tight_layout()
plt.savefig('data/sampled_signal_plots_with_trace.png') 
```
